#### Plots in Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

air_quality = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv", index_col=0, parse_dates=False)
```

```python
air_quality.head(3)
```

```python
air_quality.shape
```

```python
air_quality.dtypes
```

```python
air_quality.plot()
```

```python
air_quality['station_paris'].plot()
```

```python
air_quality.plot.scatter(x='station_london', y = 'station_paris', alpha = 0.5)
```

```python
# overview of the avaible plot methods
[
    method_name
    for method_name in dir(air_quality.plot)
    if not method_name.startswith("_")
]
```

```python
# using lambda
list(filter(lambda x : x[0] != "_" , dir(air_quality.plot)))
```

```python
air_quality.plot.box()
```

```python
axs = air_quality.plot.area(figsize=(12,4), subplots=True)
```

```python
fig, axs = plt.subplots(figsize=(12,4))
```

```python
air_quality.plot.area(ax=axs)
```

```python
axs.set_ylabel("NO$_2$ concentration")
```

```python
fig.savefig("no2_concentrations.png")
```
