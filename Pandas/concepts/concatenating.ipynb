## 1.1 Set Logic on the axes

````python
df4 = pd.DataFrame(
    {
        "B": ["B2", "B3", "B6", "B7"],
        "D": ["D2", "D3", "D6", "D7"],
        "F": ["F2", "F3", "F6", "F7"],
    },
    index=[2, 3, 6, 7],
)
````

````python
df4
````

````python
print(df1)
print("\n")
print(df4)
print("\n")
result = pd.concat([df1,df4])
print(result)
````

````python
print(df1)
print("\n")
print(df4)
print("\n")
result = pd.concat([df1,df4],axis = 1, join = 'inner')
print(result)
````

````python
print(df1)
print("\n")
print(df4)
print("\n")
result = pd.concat([df1,df4], axis = 1).reindex(df1.index)
print(result)
````

### 2.1 Brief primer on merge methods (Relational Algebra)

````python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
````

````python
right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
````

````python
print(left)
print()
print(right)
````

````python
result = pd.merge(left,right, on="key")
result
````

````python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)
````

````python
result = pd.merge(left,right,how='cross')
result
````

### 2.7 Joining a single index to a multiIndex

````python
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]},
    index=pd.Index(["K0", "K1", "K2"], name="key"),
)


index = pd.MultiIndex.from_tuples(
    [("K0", "Y0"), ("K1", "Y1"), ("K2", "Y2"), ("K2", "Y3")],
    names=["key", "Y"],
)


right = pd.DataFrame(
    {"C": ["C0", "C1", "C2", "C3"], "D": ["D0", "D1", "D2", "D3"]},
    index=index,
)


result = left.join(right, how="inner")
result
````

````python
result = pd.merge(
    left.reset_index(), right.reset_index(), on=["key"], how="inner"
).set_index(["key","Y"])
````

````python
result
````

### 2.9. MErgin on a combination of columns and index levels

````python
left_index = pd.Index(["K0", "K0", "K1", "K2"], name="key1")

left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "key2": ["K0", "K1", "K0", "K1"],
    },
    index=left_index,
)


right_index = pd.Index(["K0", "K1", "K2", "K2"], name="key1")

right = pd.DataFrame(
    {
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
        "key2": ["K0", "K0", "K0", "K1"],
    },
    index=right_index,
)


result = left.merge(right, on=["key1", "key2"])
result
````

### 2.10 Overlapping value columns

````python
left = pd.DataFrame({"k": ["K0", "K1", "K2"], "v": [1, 2, 3]})

right = pd.DataFrame({"k": ["K0", "K0", "K3"], "v": [4, 5, 6]})

result = pd.merge(left, right, on="k")

result
````

````python
result = pd.merge(left, right, on="k", suffixes=("_l", "_r"))
result
````

````python
left = left.set_index("k")

right = right.set_index("k")

result = left.join(right, lsuffix="_l", rsuffix="_r")
result
````

### 2.11 Joining Multiple DataFrames

````python
right2 = pd.DataFrame({"v": [7, 8, 9]}, index=["K1", "K1", "K2"])

result = left.join([right, right2])
result
````
